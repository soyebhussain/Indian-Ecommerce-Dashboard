```python
# ðŸ“˜ Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# âœ… Read the cleaned data
df = pd.read_csv("../data/processed/cleaned_diwali_sales_processed.csv")

# Quick check
print("Data Loaded Successfully âœ…")
print(df.head())

# ðŸŽ¯ Basic Info
print("\nDataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())

# --- Gender-wise Sales ---
gender_sales = df.groupby('gender')['amount'].sum().reset_index()
plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='amount', data=gender_sales, palette='viridis')
plt.title("Total Sales by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Sales (Amount)")
plt.show()

# --- State-wise Sales ---
state_sales = df.groupby('state')['amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
state_sales.plot(kind='bar', color='skyblue')
plt.title("Top 10 States by Total Sales")
plt.xlabel("State")
plt.ylabel("Total Sales (Amount)")
plt.xticks(rotation=45)
plt.show()

# --- Occupation-wise Sales ---
occupation_sales = df.groupby('occupation')['amount'].sum().reset_index().sort_values(by='amount', ascending=False)
plt.figure(figsize=(8,5))
sns.barplot(x='occupation', y='amount', data=occupation_sales, palette='cool')
plt.title("Total Sales by Occupation")
plt.xticks(rotation=45)
plt.show()
```
