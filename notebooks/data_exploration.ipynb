```python
# Step 1: Import libraries
import pandas as pd

# Step 2: Load the data
file_path = "../data/raw/cleaned_Diwali_sales.csv"
df = pd.read_csv(file_path)

# Step 3: View first few rows
print("Preview of data:")
display(df.head())

# Step 4: Check shape and columns
print(f"\nNumber of rows and columns: {df.shape}")
print("\nColumns in dataset:")
print(df.columns.tolist())

# Step 5: Basic info and stats
print("\nData Information:")
df.info()

print("\nSummary Statistics:")
display(df.describe())

# Step 6: Check missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Step 7: Check duplicates
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
# Step 1: Drop unnecessary column
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')


# Step 2: Rename columns (remove spaces and make lowercase)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Step 3: Handle missing values
# Fill missing 'amount' with median value (more stable than mean for skewed data)
df['amount'].fillna(df['amount'].median(), inplace=True)

# Fill missing 'age_group' based on 'age'
df['age_group'].fillna(df['age'].apply(
    lambda x: '0-18' if x <= 18 else '19-30' if x <= 30 else '31-45' if x <= 45 else '46+'
), inplace=True)

# Step 4: Verify cleaning
print("✅ Missing values after cleaning:")
print(df.isnull().sum())

# Step 5: Preview cleaned data
print("\nPreview of cleaned data:")
display(df.head())

# Step 6: Save cleaned data into processed folder
processed_path = "../data/processed/cleaned_diwali_sales_processed.csv"
df.to_csv(processed_path, index=False)

print(f"\n✅ Cleaned data saved successfully to: {processed_path}")
```
